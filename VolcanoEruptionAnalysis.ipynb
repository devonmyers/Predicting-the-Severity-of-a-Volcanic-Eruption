```python
def count_null(df):
    '''For a given dataframe, compute the number of missing values in each column of the dataframe'''
    null_counts = []
    for col in df:
        variable = {}
        count = df[col].isnull().sum()
        variable[col] = count
        null_counts.append(variable)
    return null_counts  
```

```python
# Create a dictionary for number of eruptions.  Dictionary keys are volcano numbers, values are counts of the number of eruptions
volcano_nums = df1['Volcano Number'].values
num_eruptions_dict = {}
for num in volcano_nums:
    if num not in num_eruptions_dict: # volcano number has not been seen yet, set the value of this volcano in the dict to 1
        num_eruptions_dict[num] = 1
    else:
        count = num_eruptions_dict[num]
        count += 1
        num_eruptions_dict[num] = count
```
